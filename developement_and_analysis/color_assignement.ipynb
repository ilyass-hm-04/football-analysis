```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
image_path = '../output_videos/croped_image.jpg'
image = cv2.imread(image_path)
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show()
```

# Take top half image

```python
top_half_image = image[0:int(image.shape[0]//2),:]
plt.imshow(top_half_image)
plt.show()
```

# Cluster the image into two clusters

```python
image2d = top_half_image.reshape(-1,3)

kmeans = KMeans(n_clusters=2,random_state=0).fit(image2d)

labels = kmeans.labels_

clustured_image = labels.reshape(top_half_image.shape[0],top_half_image.shape[1])

plt.imshow(clustured_image)
plt.show()
```

```python
corners_cluster = [clustured_image[0,0],clustured_image[0,-1],clustured_image[-1,0],clustured_image[-1,-1]]
non_payer_cluster = max(set(corners_cluster),key=corners_cluster.count)
print(non_payer_cluster)
```

```python
player_cluster = 1-non_payer_cluster
print(player_cluster)
```

```python
kmeans.cluster_centers_[player_cluster]
```
